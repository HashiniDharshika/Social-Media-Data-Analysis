## Step 1: Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

## Step 2: Generate Random Data for Social Media

```python
categories = ['Food', 'Travel', 'Fashion', 'Fitness', 'Music', 'Culture', 'Family', 'Health']
n = 500

# Generate random data
data = {
    'Date': pd.date_range('2022-12-30', periods=n),
    'Category': [random.choice(categories) for _ in range(n)],
    'Likes': np.random.randint(0, 10000, size=n)
}
```

```python
# Convert the data dictionary to a DataFrame
df = pd.DataFrame(data)
```

## Step 3: Load and Explore the Data

```python
# Display the first few rows of the DataFrame
print(df.head(20))
```

```python
# Display DataFrame information
print(df.info())

# Display DataFrame description
print(df.describe())

# Count of each Category element
print(df['Category'].value_counts())
```

## Step 4: Export the database

```python
# Export the DataFrame to a CSV file
df.to_csv(r'C:\Users\User\Clean and analyze social media usage data with Python\Output database\social_media_data.csv', index=False)
```

## Step 5: Clean the Data


```python
# Remove any null data
df.dropna(inplace=True)

# Remove duplicate data
df.drop_duplicates(inplace=True)

# Convert 'Date' to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Ensure 'Likes' are integers
df['Likes'] = df['Likes'].astype(int)

print(df.info())  # Verify changes
```

## Step 5:Visualize and Analyze the Data

```python
# Group the data by category and sum the likes
likes_by_category = df.groupby('Category')['Likes'].sum().reset_index()

# Plot a bar chart of the total likes by category
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Likes', data=likes_by_category, palette='viridis')
plt.title('Total Likes Count by Category')
plt.xlabel('Category')
plt.ylabel('Total Likes')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Category', y='Likes', data=df)
plt.title('Likes by Category')
plt.xlabel('Category')
plt.ylabel('Likes')
plt.xticks(rotation=45)
plt.show()
```

```python
# Mean of 'Likes'
mean_likes = df['Likes'].mean()
print(f"Overall mean of Likes: {mean_likes}")

# Mean of 'Likes' per Category
mean_likes_by_category = df.groupby('Category')['Likes'].mean()
print("Mean of Likes by Category:")
print(mean_likes_by_category)
```

## Time Series Analysis

```python
# Load the data from the CSV file
file_path = (r'C:\Users\User\Clean and analyze social media usage data with Python\Output database\social_media_data.csv')
df = pd.read_csv(file_path)

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Group the data by 'Date' and 'Category', and sum the likes
grouped_data = df.groupby(['Date', 'Category'])['Likes'].sum().reset_index()

# Filter the data for only 'Family' and 'Food' categories
filtered_data = grouped_data[grouped_data['Category'].isin(['Family', 'Food'])]

# Time Series Analysis for 'Family' and 'Food' categories
plt.figure(figsize=(14, 8))
sns.set(style='whitegrid')
sns.lineplot(data=filtered_data, x='Date', y='Likes', hue='Category', marker='o')
plt.title('Likes Trend Over Time for Family and Food Categories')
plt.xlabel('Date')
plt.ylabel('Total Likes')
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

## Correlation Analysis

```python
# Pivot the data to get categories as columns
pivot_df = df.pivot_table(values='Likes', index='Date', columns='Category', aggfunc='sum').fillna(0)

# Calculate the correlation matrix
correlation_matrix = pivot_df.corr()

# Plot the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Likes Between Categories')
plt.show()
```
